```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../df/CleanFoodData.csv")
data = data[["Country", "Pop_Class","Food", "Mean"]]
data['Z'] = [0]*len(data)
data
```

```python
data.loc[1:2,"Mean"]
```

```python
countries = list(data['Country'].unique())
ageGroup = list(data['Pop_Class'].unique())
indexes=[];
total_group = 0
for country in countries:
    for age in ageGroup:
        for i, row in data.iterrows():
            if row['Country'] == country and row['Pop_Class'] == age:
                indexes.append(i)
                total_group = total_group + row['Mean']
        data.loc[indexes, 'Z'] = total_group
        indexes = []
        total_group = 0
data
```

```python
data['Mean_norm'] = data['Mean']/data['Z']
data.head(70)
```

```python
data.to_csv("./normalized_food_per_country_ageGroup.csv", header=True, sep=',', index=False)
```

```python
countries = list(data['Country'].unique())
ageGroup =list(data['Pop_Class'].unique())
cols = ["Country", "ageGroup"] + list(set(data['Food']))
data_format = pd.DataFrame([[0]*(len(set(data['Food']))+2)]* len(data['Pop_Class'].unique()) * len(data['Country'].unique()), columns=cols)
i = 0
for country in countries:
    for age in ageGroup:
        data_format.loc[i,'Country'] = country
        data_format.loc[i,'ageGroup'] = age
        i+=1  
data_format
```

```python
for i, row in data_format.iterrows():
    for iRaw, rowRaw in data.iterrows():
        if row['Country'] == rowRaw['Country'] and row['ageGroup'] == rowRaw['Pop_Class']:
            for food in data_format.columns.values:
                if food == data.loc[iRaw,'Food']:
                    data_format.loc[i, food] = data.loc[iRaw,'Mean_norm']
data_format
```

```python
data_format.to_csv("./formated_normalized_food_per_country_ageGroup.csv", sep=",", header=True, index=False)
```
