```python
import numpy as np
from sklearn.preprocessing import Imputer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
import pandas as pd
import random
```

```python
data_csv = pd.read_csv("../../data/formated_normalized_food_per_country_ageGroup.csv")
data_csv
```

```python
X = np.asarray(data_csv.iloc[:,2:])
y_country = np.asarray(data_csv.iloc[:,0])
y_age = np.asarray(data_csv.iloc[:,1])
```

```python
neigh_country = KNeighborsClassifier(n_neighbors = 1, weights='uniform', algorithm='auto')
neigh_country.fit(X, y_country)
neigh_age = KNeighborsClassifier(n_neighbors = 1, weights='uniform', algorithm='auto')
neigh_age.fit(X, y_age);
```

```python
# Testing
noise = np.random.normal(0,1,(len(data_csv.columns) - 2)) * 0.05
i = random.randint(0,len(data_csv))
y_pred_country = neigh_country.predict([X[i] + noise])
print('Predicted country: ' + y_pred_country[0])
print('Actual country: ' + y_country[i])
```

```python
# Testing
noise = np.random.normal(0,1,(len(data_csv.columns) - 2)) * 0.05
i = random.randint(0,len(data_csv))
y_pred_age = neigh_age.predict([X[i] + noise])
print('Predicted age group: ' + y_pred_age[0])
print('Actual country: ' + y_age[i])
```
